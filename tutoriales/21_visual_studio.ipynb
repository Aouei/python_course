# Visual Studio

Visual Studio es un Entorno de desarrollo integrado. Es decir, es una herramienta compuesta por varias herramientas bien integradas entre sí con el fin de facilitar la tarea de desarrollo.

Entre sus componentes destacamos:
- Explorador
- Búsqueda
- Control de versiones
- Extensiones
- Terminal
- Editor de texto / Visualizador de archivos

## Atajos

Panel:
- Ctrl + ñ | Ctrl + j -> Abre una nueva terminal
- Ctrl + b -> Abre o cierra el panel izquierdo (Explorador, Búsqueda, Control de versiones, ...)
- Ctrl + Shift + E -> Abre el Explorador
- Ctrl + Shift + F -> Abre la Búsqueda
- Ctrl + Shift + G -> Abre el Control de versiones
- Ctrl + Shift + X -> Abre el Gestor de extensiones

Escritura:
- Shift + Up Arrow -> Mueve linea o lineas seleccionadas hacia arriba
- Shift + Down Arrow -> Mueve linea o lineas seleccionadas hacia abajo
- Ctrl + Alt + Down Arrow -> Crea otro curso hacia abajo
- Ctrl + Alt + Up Arrow -> Crea otro curso hacia arriba
- Shift + Alt + Down Arrow -> Copia la linea o lineas seleccionadas hacia abajo
- Shift + Alt + Up Arrow -> Copia la linea o lineas seleccionadas hacia arriba
- Ctrl + D -> Busca y selecciona el mismo patrón que actualmente tienes seleccionado
- Ctrl + K -> Elimina el último elemento seleccionado por Ctrl + D
